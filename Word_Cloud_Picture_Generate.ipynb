```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample text data
text = "Ujjal Mandal Matlab Computational Physics Research Scholar Python C Quantum Mechanics Classical Mechanics Comsol Biomedical Data Science Electromagnetic Acoustics Sound CUDA CSS HTML Java script JAVA C++ Fortran"

# Generate the word cloud with LinkedIn banner size
wordcloud = WordCloud(width=1584, height=396, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(15.84, 3.96), facecolor=None)  # Adjust figsize to match the dimensions
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Turn off axis
plt.tight_layout(pad=0)

# Save the word cloud as a PNG file without any border
plt.savefig('linkedin_banner_wordcloud.png', dpi=300, bbox_inches='tight', pad_inches=0)

# Show the word cloud
plt.show()
```
